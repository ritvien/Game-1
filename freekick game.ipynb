```python
from tkinter import * 
from PIL import ImageTk, Image
from time import sleep
import random
import tkinter.font as tkFont
class Click:
    def instructor(e):
        btn_play.place(x=900,y=400)
        btn_instr.place(x=900,y=400)
        canvas.move(instruc,-600,0)
        btn_back.place(x=240,y=550)
    def back(e):
        btn_play.place(x=220,y=400)
        btn_instr.place(x=220,y=450)
        canvas.move(instruc,600,0)
        btn_back.place(x=940,y=550)
    def play(e):
        btn_play.place(x=900,y=400)
        btn_instr.place(x=900,y=400)
        lb_level1.place(x=0,y=50)
        canvas.move(first,600,0)
        canvas.bind_all('<KeyPress-Up>', go_up)
        canvas.bind_all('<KeyPress-Down>', go_down)
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
        while 1:
            move()
            canvas.update_idletasks()
            canvas.update()
    def Tryagain_goal(e):
        global run_y
        pos1 = canvas.coords(cr7)
        btn_nextlv.place(x=900,y=650)
        btn_nextlv2.place(x=900,y=650)
        btn_tryagain_goal.place(x=900,y=650)
        canvas.move(cele,600,0)
        canvas.move(cr7,-62,pos1[1]-350)
        pos2 = canvas.coords(ball)
        canvas.move(ball,150-pos2[0],400-pos2[1])
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
    def Tryagain_miss(e):
        global run_y
        pos1 = canvas.coords(cr7)
        btn_tryagain_miss.place(x=900,y=650)
        canvas.move(miss,600,0)
        canvas.move(cr7,-62,pos1[1]-350)
        pos2 = canvas.coords(ball)
        canvas.move(ball,150-pos2[0],400-pos2[1])
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
    def Tryagain_catch(e):
        global run_y
        pos1 = canvas.coords(cr7)
        btn_tryagain_catch.place(x=900,y=650)
        canvas.move(catch,600,0)
        canvas.move(cr7,-62,pos1[1]-350)
        pos2 = canvas.coords(ball)
        canvas.move(ball,150-pos2[0],400-pos2[1])
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
    def Tryagain_bar_catch(e):
        global run_y
        pos1 = canvas.coords(cr7)
        btn_tryagain_bar_catch.place(x=900,y=650)
        canvas.move(bar_catch,600,0)
        canvas.move(cr7,-62,pos1[1]-350)
        pos2 = canvas.coords(ball)
        canvas.move(ball,150-pos2[0],400-pos2[1])
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
    def Next_level(e):
        global run_y
        pos1 = canvas.coords(cr7)
        btn_nextlv.place(x=900,y=650)
        btn_tryagain_goal.place(x=900,y=650)
        lb_level2.place(x=0,y=50)
        canvas.move(cele,600,0)
        canvas.move(cr7,-62,pos1[1]-350)
        pos2 = canvas.coords(ball)
        canvas.move(ball,150-pos2[0],400-pos2[1])
        canvas.move(bar,-500,30)
        #barx : [373;427] bary : [355;505] 
        canvas.move(gk,0,-45)
        global current_lv
        current_lv += 1
        canvas.bind_all('<Return>',shoot)
        canvas.bind_all('<Shift-Return>',curl_shoot)
    def Next_level2(e):
        canvas.move(update,-600,0)
        btn_nextlv2.place(x=900,y=650)
        btn_tryagain_goal.place(x=900,y=650)
        lb_level3.place(x=0,y=50)
        
        
#Cr7:
y = 0
current_lv = 1

def blank(e):
    pass
def go_up(evt):
    global y
    y = -2
def go_down(evt):
    global y
    y = 2
def move():
    global y
    canvas.move(cr7,0,y)
    y = 0
def shoot(evt):
    pos = canvas.coords(cr7)
    global run_y
    run_y = (350-pos[1])/33
    global spin
    spin = 0   
    for i in range(1,32):
        canvas.move(cr7,2,run_y)
        canvas.update()
        sleep(0.002)
    canvas.bind_all('<Return>',blank)
    canvas.bind_all('<Shift-Return>',blank)
    straight()
def curl_shoot(evt):
    pos = canvas.coords(cr7)
    global run_y
    run_y = (350-pos[1])/33
    for i in range(1,32):
        canvas.move(cr7,2,run_y)
        canvas.update()
        sleep(0.002)
    global spin
    spin = 0
    if run_y<=0:
        spin = 0.07
    elif run_y>0:
        spin = -0.07
    canvas.bind_all('<Return>',blank)
    canvas.bind_all('<Shift-Return>',blank)
    curl()


#BALL
def straight():
    for i in range(1,105):
        canvas.move(ball,4,run_y)
        canvas.update()
        sleep(0.0005)
        #check_goal
        pos_ball = canvas.coords(ball)
        if 373 < pos_ball[0] < 427 and 372.5 < pos_ball[1] < 487.5 and  current_lv == 2:
            goal2()
            break
        elif pos_ball[0] == 566 and current_lv == 1:
            goal()
        elif pos_ball[0] == 566 and current_lv == 2:
            goal2()
def curl():
    global run_y
    global spin
    for i in range(1,45):
        canvas.move(ball,4,run_y)
        canvas.update()
        sleep(0.0005)
        #check_goal
        pos_ball = canvas.coords(ball)
        if 373 < pos_ball[0] < 427 and 372.5 < pos_ball[1] < 487.5 and  current_lv == 2:
            goal2()
            break
        elif pos_ball[0] == 566 and current_lv == 1:
            goal()
        elif pos_ball[0] == 566 and current_lv == 2:
            goal2()
    for i in range(45,105):
        canvas.move(ball,4,run_y)
        canvas.update()
        sleep(0.0005)
        run_y = run_y + spin
        #check_goal
        pos_ball = canvas.coords(ball)
        if 373 < pos_ball[0] < 427 and 372.5 < pos_ball[1] < 487.5 and  current_lv == 2:
            goal2()
            break
        elif pos_ball[0] == 566 and current_lv == 1:
            goal()
        elif pos_ball[0] == 566 and current_lv == 2:
            goal2()
def goal():
    #Goaly = [330;470] #GKy = [362.5;437.5]
    pos_ball = canvas.coords(ball)
    if  pos_ball[1] < 330 or pos_ball[1] > 470:
        canvas.move(miss,-600,0)
        btn_tryagain_miss.place(x=80,y=650)
    elif 362.5 < pos_ball[1] <437.5:
        canvas.move(catch,-600,0)
        btn_tryagain_catch.place(x=80,y=650)
    else:    
        canvas.move(cele,-600,0)
        btn_nextlv.place(x=400,y=650)
        btn_tryagain_goal.place(x=80,y=650)
def goal2():
    #GKy = [317.5;392.5]
    #barx : [373;427] bary : [372.5;487.5] 
    pos_ball = canvas.coords(ball)
    if 373 < pos_ball[0] < 427 and 372.5 < pos_ball[1] < 487.5:
        canvas.move(bar_catch,-600,0)
        btn_tryagain_bar_catch.place(x=80,y=650)
    elif pos_ball[0]==566 and 317.5 < pos_ball[1] < 392.5 :
        canvas.move(catch,-600,0)
        btn_tryagain_catch.place(x=80,y=650)
    elif pos_ball[0]==566 and (pos_ball[1] < 330 or pos_ball[1] > 470) :
        canvas.move(miss,-600,0)
        btn_tryagain_miss.place(x=80,y=650)
    elif pos_ball[0] ==566:
        canvas.move(cele,-600,0)
        btn_nextlv2.place(x=400,y=650)
        btn_tryagain_goal.place(x=80,y=650)
        

    
        
                    

            
root = Tk()
root.title("Ronaldo Freekick")
root.geometry('600x800')
root.resizable(0,0)
click = Click()

#Tạo hình:
canvas= Canvas(root, width= 600, height= 800)
canvas.pack()


#Image
first_img = ImageTk.PhotoImage(Image.open('First.png'))
field_img = ImageTk.PhotoImage(Image.open('bg.png'))
ball_img = ImageTk.PhotoImage(Image.open('ball.png'))
cr7_img = ImageTk.PhotoImage(Image.open('cr7.jpg'))
gk_img = ImageTk.PhotoImage(Image.open('gk.png'))
bar_img = ImageTk.PhotoImage(Image.open('barrier.png'))
bar_catch_img = ImageTk.PhotoImage(Image.open('wall.png'))
cele_img = ImageTk.PhotoImage(Image.open('goal.jpg'))
miss_img = ImageTk.PhotoImage(Image.open('miss.jpg'))
instruc_img = ImageTk.PhotoImage(Image.open('Instruc.png'))
catch_img = ImageTk.PhotoImage(Image.open('catch.png'))
update_img = ImageTk.PhotoImage(Image.open('update.png'))

bg_field = canvas.create_image(300,400,image=field_img)
ball = canvas.create_image(150,400,image=ball_img)
cr7 = canvas.create_image(50,350,image=cr7_img)
gk = canvas.create_image(550,400,image=gk_img)
bar = canvas.create_image(900,400,image=bar_img)
bar_catch = canvas.create_image(900,400,image=bar_catch_img)
cele = canvas.create_image(900,400,image=cele_img)
miss = canvas.create_image(900,400,image=miss_img)
catch = canvas.create_image(900,400,image=catch_img)
first = canvas.create_image(300,400,image=first_img)
instruc = canvas.create_image(900,400,image=instruc_img)
update = canvas.create_image(900,400,image=update_img)


#Button, Label
btn_play = Button(root, text = "Play", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 15,command=click.play)
btn_instr = Button(root, text = "Instructor", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 15,command=click.instructor)
btn_back = Button(root, text = "Back", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.back)
btn_nextlv = Button(root, text = "Next level", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Next_level)
btn_nextlv2 = Button(root, text = "Next level", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Next_level2)
btn_tryagain_goal = Button(root, text = "Try Again", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Tryagain_goal)
btn_tryagain_miss = Button(root, text = "Try Again", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Tryagain_miss)
btn_tryagain_catch = Button(root, text = "Try Again", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Tryagain_catch)
btn_tryagain_bar_catch = Button(root, text = "Try Again", bg='sky blue', fg='black', activeforeground = "red", activebackground = "orange",bd=2,font="-weight bold" ,width = 10,command=click.Tryagain_bar_catch)
lb_level1 = Label(root, text = "Level 1",bg='sky Blue',fg='white',bd=2,font="-weight bold",width = 7)
lb_level2 = Label(root, text = "Level 2",bg='sky Blue',fg='white',bd=2,font="-weight bold",width = 7)
lb_level3 = Label(root, text = "Level 3",bg='sky Blue',fg='white',bd=2,font="-weight bold",width = 7)
btn_play.place(x=220,y=400)
btn_instr.place(x=220,y=450)



root.mainloop()
```
